```python
     
def stream_tweets(search_term):
    tweets_list = []
    counter = 0
    for tweet in tweepy.Cursor(api.search, q='\"{}\" -filter:retweets'.format(search_term), count=166666, lang='en', tweet_mode='extended').items():
        tweet_details = {}
        tweet_details['user_name'] = tweet.user.screen_name
        tweet_details['user_location'] = tweet.user.location
        tweet_details['date'] = tweet.user.created_at.strftime("%d-%b-%Y")

        tweet_details['text']= tweet.full_text

        if hasattr(tweet, 'place'):
            try:
                tweet_details['country_code'] = tweet.place['country_code']
                tweet_details['place_name'] = tweet.place['full_name']
                tweet_details['place_type'] = tweet.place['place_type']
                tweet_details['location_coordinates'] = tweet.place['bounding_box']['coordinates']
            except:
                pass
        else:#in order to avoid error : local variable 'country_code' referenced before assignment; 
                    #sometimes we dont have the value, and these variables will only be local variables, which cannot be returned as global variables.
            tweet_details['country_code'] = ''
            tweet_details['place_name'] = ''
            tweet_details['place_type'] = ''
            tweet_details['location_coordinates'] = ''
                    
        tweets_list.append(tweet_details)
                
        counter += 1
        if counter == 166666:
            break
        else:
            pass
    with open('tweets_{}.json'.format(search_term), 'a+') as f:
        json.dump(tweets_list, f)
    print('done!')
                
```

```python
search_terms = ['COVID-19','pandemic','coronavirus']
for search_term in search_terms:
        stream_tweets(search_term)
```

```python
import pandas as pd
import numpy as np
import json
import re
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)', ' ', tweet).split())
```

```python
tweets1 = pd.read_json('tweets_COVID-19.json', orient='records', dtype={"A":str, "B":str})
tweets1['text'] = tweets1['text'].apply(lambda x: clean_tweet(x))
```

```python
tweets3 = pd.read_json('tweets_pandemic.json', orient='records', dtype={"A":str, "B":str})
tweets3['text'] = tweets3['text'].apply(lambda x: clean_tweet(x))
```

```python
tweets2 = pd.read_json('tweets_coronavirus.json', orient='records', dtype={"A":str, "B":str})
tweets2['text'] = tweets2['text'].apply(lambda x: clean_tweet(x))
```
